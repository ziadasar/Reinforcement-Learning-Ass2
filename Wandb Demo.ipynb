## Used Libraries

```python
import torch
import torch.nn as nn
import torch.optim as optim
import wandb
import os
```

## Get Api Key From a File and Initialize your Project

```python
with open("key.txt", "r") as f:
    api_key = f.read().strip()
wandb.login(key=api_key)
wandb.init(project="wandb_demo_project", name="dummy_training_run") # project name and run name
```

## Dummy NN for Testing

```python
# Dummy dataset (100 samples, 10 features each)
X = torch.randn(100, 10)
y = torch.randint(0, 2, (100,))  # binary classification
```

```python
class DummyNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, 2)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)
```

```python
model = DummyNet()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
wandb.watch(model, log="all")
# Dummy training loop
for epoch in range(10000):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()

    # Compute fake accuracy
    preds = torch.argmax(outputs, dim=1)
    acc = (preds == y).float().mean().item()

    # Log metrics to W&B
    wandb.log({"epoch": epoch + 1, "loss": loss.item(), "accuracy": acc})

    print(f"Epoch [{epoch+1}/10] - Loss: {loss.item():.4f} | Accuracy: {acc:.4f}")
```

```python
wandb.finish()
```
